#  Import Libraries

```python
# Importing Libraries
import requests as req
import json
import pandas as pd
```

# Create Query URL  and a empty list to store cities

### Query URL is the used for getting information from the API

```python
# Getting base URL for Open Weather API
base_url= 'http://api.openweathermap.org/data/2.5/weather'

# Getting API key
key= '2cd240c12582cb116b1eeb99619e4091'

# Creating an empty list to store our JSON response objects
weather_data = []

# Defining the multiple cities we would like to make a request for
cities = ["Kalyan", "Mumbai", "Chennai", "Delhi", "Kolkata", "New York","Los Angeles","Chicago", "Dallas"]

# Reading through each city in our cities list and performing a request call to the API.
# Storing each JSON response object into the list
for city in cities:
    query_url = base_url + "?apikey=" + key + "&q=" + city
    weather_data.append(req.get(query_url).json())
```

# Data in JSON Format

```python
weather_data
```

# Extract data for multiple cities and storing it into Pandas Dataframe

```python
# Creating an empty list for each variable
city_name = []
temperature_data = []
weather_description_data = []

# Extracting the city name, temperature, and weather description of each City
for data in weather_data:
    city_name.append(data['name'])
    temperature_data.append(data['main']['temp'])
    weather_description_data.append(data['weather'][0]['description'])

# Printing out the list to make sure the queries were extracted 
print ("The City Name List: " + str(city_name))
print ("The Temperature List: " + str(temperature_data))
print ("The Weather Description List: " + str(weather_description_data))
```

# Create Dictionary to store Data

```python
# Create a dictionary containing our newly extracted information
weather_data = {"City": city_name, 
                "Temperature": temperature_data,
                "Weather Description": weather_description_data}
```

# Convert Dictionary to Pandas DataFrame

```python
# Converting our dictionary into a Pandas DataFrame
weather_data = pd.DataFrame(weather_data).reset_index()
weather_data
```
